## 1. Tipe data


<h2>1.1 Atribut Nominal / Kategorical</h2>


<p>Nominal berarti "berkaitan dengan nama." Nilai atribut nominal adalah simbol atau nama benda. Setiap nilai mewakili beberapa jenis kategori, kode, atau status, sehingga atribut nominal juga disebut sebagai kategori. Nilai tidak memiliki urutan yang berarti. Dalam ilmu komputer, nilai juga dikenal sebagai pencacahan.</p>

<h2>1.2 Atribut Binary (Biner)</h2>

<p>Atribut biner adalah atribut nominal dengan hanya dua kategori atau status 0 atau 1, di mana 0 biasanya berarti bahwa atribut tersebut tidak ada, dan 1 berarti ada. Atribut biner disebut sebagai Boolean jika kedua status sesuai dengan benar dan salah.</p>

<h2>1.3 Ordinal</h2>

<p>Atribut ordinal adalah atribut dengan kemungkinan nilai yang memiliki urutan atau peringkat yang bermakna di antara mereka, tetapi besarnya antara nilai-nilai yang berurutan tidak diketahui.</p>

<h2>1.4 Numeric</h2>

<p>Atribut numerik bersifat kuantitatif; yaitu, itu adalah kuantitas yang dapat diukur, direpresentasikan dalam bilangan bulat atau nilai nyata. Atribut numerik dapat berskala interval atau berskala rasio.</p>

<h2>1. Menghitung Jarak Dissimilarity Atribut Binary</h2>

*   Ambil Sampel pada ID Pasien dan Kemungkinan terkena kanker payudara

```python
data_binary = df[["ID", "Kemungkinan Terkena Kanker Payudara"]]

data_binary
```



*   Source Code


Parameters:

*   id (str): ID pasien pertama.
*   id2 (str): ID pasien kedua.
*   attribute (str): nama atribut yang akan dihitung jarak dissimilarity-nya.



```python
df = pd.read_csv("https://gist.githubusercontent.com/iamwilldev/2a73758f9e4e98b71097715c6043381c/raw/a626cfa9d1b61a4b66aee45ca48af0a6efdcffb1/medis-pd-dataset.csv")

# ganti nilai pada kolom "Kemungkinan Terkena Kanker Payudara" dengan 0 dan 1
df['Kemungkinan Terkena Kanker Payudara'] = df['Kemungkinan Terkena Kanker Payudara'].map({'Tidak': 0, 'Ya': 1})

# Ambil data hanya dari kolom "ID" dan "Kemungkinan Terkena Kanker Payudara"
data_binary = df[["ID", "Kemungkinan Terkena Kanker Payudara"]]

def calculate_dissimilarity_distance(id, id2, attribute):
    # cek apakah id ada di tabel
    if id not in data_binary['ID'].values:
        print(f"The ID {id} does not exist in the dataset.")
    else:
        # mendapatkan nilai untuk atribut yang ditentukan dari ID yang diberikan
        value1 = data_binary.loc[data_binary['ID'] == id, attribute].values[0]

        # cek apakah id ada di tabel
        if id2 not in data_binary['ID'].values:
            print(f"The ID {id2} does not exist in the dataset.")
        else:
            # mendapatkan nilai untuk atribut yang ditentukan dari ID acak
            value2 = data_binary.loc[data_binary['ID'] == id2, attribute].values[0]

            # Menghitung jarak dissimilarity
            r = int(value1 != value2 and value1 == 1)
            s = int(value1 != value2 and value2 == 1)
            q = int(value1 == value2 and value1 == 0)
            t = int(value1 == value2 and value1 == 1)
            d = (r + s) / (q + r + s + t)

            # menampilkan hasil
            print(f"Jarak dissimilarity Binary antara {id} dan {id2} pada kolom {attribute} adalah: {d:.2f}")

# Call the function with different inputs to demonstrate its use
calculate_dissimilarity_distance('P001', 'P002', 'Kemungkinan Terkena Kanker Payudara')
calculate_dissimilarity_distance('P001', 'P003', 'Kemungkinan Terkena Kanker Payudara')
```

<h2>2. Menghitung Jarak Dissimilarity Atribut Numeric</h2>

*   Ambil Sampel pada ID Pasien, Umur, Ukuran tumor dan Jumlah nodus getah bening

```python
data_numeric = df[["ID", "Umur", "Ukuran Tumor (cm)", "Jumlah Nodus Getah Bening"]]

data_numeric
```



*   Source Code


Parameters:

*   id1 (str): ID pasien pertama.
*   id2 (str): ID pasien kedua.
*   atribut[]: Umur, Ukuran Tumor (cm), Jumlah Nodus Getah Bening
*   data_numeric: Tabel data frame



```python
# Fungsi untuk menghitung jarak Euclidean antara dua data
def euclidean_distance(atribut1, atribut2):
    jarak = 0
    for i in range(len(atribut1)):
        jarak += (atribut1[i] - atribut2[i]) ** 2
    jarak = math.sqrt(jarak)
    return jarak

# Fungsi untuk menghitung jarak dissimilarity antara dua pasien
def hitung_jarak(id1, id2, atribut, data_numeric):
    # Ambil data atribut dari kedua pasien
    atribut1 = [data_numeric.loc[data_numeric['ID'] == id1, kolom].values[0] for kolom in atribut]
    atribut2 = [data_numeric.loc[data_numeric['ID'] == id2, kolom].values[0] for kolom in atribut]
    
    # Hitung jarak Euclidean antara kedua pasien
    jarak = euclidean_distance(atribut1, atribut2)
    
    return jarak

jarak = hitung_jarak('P001', 'P003', ["Umur", "Ukuran Tumor (cm)", "Jumlah Nodus Getah Bening"], data_numeric)
print("Jarak dissimilarity Numeric P001 dan P003 pada atribut Umur, Ukuran Tumor (cm), Jumlah Nodus Getah Bening adalah", jarak)
```

<h2>3. Menghitung Jarak Dissimilarity Atribut Nominal/Kategorical</h2>

*   Ambil Sampel pada ID Pasien dan Jenis Kanker Payudara

```python
data_kategorical = df[["ID", "Jenis"]]

data_kategorical
```



*   Source Code


Parameters:

*   data (DataFrame): dataset yang berisi data pasien.
*   attribute (str): nama atribut yang akan dihitung jarak dissimilarity-nya.
*   patient1 (str): ID pasien pertama.
*   patient2 (str): ID pasien kedua.



```python
import random

def simple_matching_distance(data, attribute, patient1, patient2):

    # Mencari jumlah variabel total pada atribut yang dianalisis
    p = len(data[attribute].unique())
    
    # Mencari jumlah variabel yang sesuai antara kedua pasien pada atribut yang dianalisis
    m = sum(data.loc[data["ID"] == patient1, attribute].iloc[0] == value for value in data[attribute].unique())
    
    # Menghitung jarak dissimilarity dengan metode simple matching
    distance = (p-m)/p
    
    return distance

# memilih dua pasien secara acak dari DataFrame table
patient1, patient2 = random.sample(list(data_kategorical["ID"]), 2)

# memilih satu atribut secara acak dari DataFrame table
attribute = 'Jenis'

# menghitung jarak dissimilarity antara dua pasien pada atribut yang dipilih dengan fungsi simple_matching_distance
distance = simple_matching_distance(data_kategorical, attribute, patient1, patient2)

# menampilkan hasil
print("Jarak dissimilarity Kategorical", patient1, "dan", patient2, "pada atribut", attribute, "Kanker adalah", distance)
```

<h2>4. Menghitung Jarak Dissimilarity Atribut Ordinal</h2>

*   Ambil Sampel pada ID Pasien dan Skor Gleason

```python
data_ordinal = df[["ID", "Skor Gleason"]]

data_ordinal
```



*   Source Code


Parameters:

*   id1 (str): ID pasien pertama.
*   id2 (str): ID pasien kedua.



```python
# buat tabel dengan data Skor Gleason
data_ordinal = df.loc[:, ["ID", "Skor Gleason"]].copy()

# fungsi untuk menghitung jarak dengan metode variabel skala interval
def calculate_distance(id1, id2):
    # ubah Skor Gleason menjadi variabel interval
    data_ordinal['Midrange Value'] = (data_ordinal['Skor Gleason'] + 1) / 2

    # ambil nilai Skor Gleason dan Midrange Value untuk kedua ID
    x = data_ordinal.loc[data_ordinal['ID'] == id1, 'Midrange Value'].values
    y = data_ordinal.loc[data_ordinal['ID'] == id2, 'Midrange Value'].values

    # hitung jarak Euclidean antara dua titik
    d = math.sqrt(sum((x - y) ** 2))

    print(f"Jarak dissimilarity Ordinal antara {id1} dan {id2} pada atribut Skor Gleason: {d}")
    
# contoh pemanggilan fungsi
calculate_distance('P001', 'P002')
calculate_distance('P001', 'P004')
```
